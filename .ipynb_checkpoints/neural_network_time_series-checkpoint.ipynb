# Predicting Time Series with a Neural Network

Time series data represents information collected over time, often at consistent intervals. This type of data is prevalent in many domains, including finance, healthcare, weather forecasting, and transportation. Analyzing time series data involves understanding its patterns, trends, seasonality, and noise to make predictions about future values. Neural networks, particularly recurrent neural networks (RNNs) and their variants like LSTMs (Long Short-Term Memory) and GRUs (Gated Recurrent Units), have proven to be powerful tools for modeling time series data due to their ability to capture temporal dependencies and learn complex patterns.

In this notebook, we will build a neural network model to forecast future values of a time series dataset: the Monthly International Air Passengers dataset, which records the number of airline passengers in thousands over a 12-year period (1949-1960). The dataset exhibits both trend and seasonality, making it an excellent candidate for time series analysis.

We will follow these steps:

1. Data Loading and Visualization: Import the dataset, plot the time series, and explore its characteristics.
2. Data Preprocessing: Prepare the data for neural network modeling by scaling, windowing, and creating training/test splits.
3. Model Development: Build and train an LSTM-based model for time series forecasting.
4. Model Evaluation and Forecasting: Evaluate the model's performance and use it to forecast future values.

This approach will showcase how neural networks can effectively handle time series data, paving the way for more complex and real-world applications.



```python
# Packages to import we will be needing to solve this exercise

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
```

```python
# Set random seeds for reproducibility
np.random.seed(777)  # For NumPy
tf.random.set_seed(777)  # For TensorFlow/Keras (includes randomness in model weights and training)
```

Before modeling, it’s essential to load and visualize the data. Visualization helps us understand the patterns, trends, seasonality, and any potential anomalies in the dataset. This step lays the groundwork for preprocessing and model development.

```python
# Load the AirPassengers dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
data = pd.read_csv(url, header=0, parse_dates=[0], index_col=0)

# Plot the data
data.plot()
plt.title('Monthly International Air Passengers')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.show()
```

Taking a look at the plot, we understand better some characteristics of our data.

- Univariate: The dataset contains a single variable: the number of airline passengers per month.

- Non-Stationary: The dataset exhibits a positive trend, meaning the values increase over time, making the time series non-stationary. The mean is not constant, and the variance may increase as the values grow.

- Seasonal: There is a seasonal pattern present in the dataset, with recurring fluctuations each year. This suggests that certain months have higher or lower passenger counts than others, forming a repeating cycle.

- Trend: The time series shows an upward trend over time, with the number of passengers generally increasing from year to year.

- Cyclicality: While the series has clear seasonal patterns (e.g., peaks in certain months), it may also exhibit cyclic fluctuations that could occur over periods longer than a year, though these are less pronounced compared to seasonality.


Our next step is to prepare the data. Neural networks require input data to be normalized or scaled for stable training. Additionally, time series models like LSTMs need sequential data as input, which we prepare using a sliding window approach, moving a fixed-size window across time series data, analyzing each subset of data as the window shifts step-by-step.

To do: 
- Normalization: Scale the data to a range (e.g., 0 to 1) to improve training stability.
- Train-Test Split: Reserve the last 20% of the data for testing.
- Time Windowing: Create supervised learning inputs (X) and targets (y) by generating windows of sequential data.

```python
# Normalization
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data.values)

# Train-test split
train_size = int(len(scaled_data) * 0.8)
train_data, test_data = scaled_data[:train_size], scaled_data[train_size:]

# Time Windowing
def create_dataset(dataset, look_back=12):
    X, y = [], []
    for i in range(len(dataset) - look_back):
        X.append(dataset[i:i + look_back, 0])
        y.append(dataset[i + look_back, 0])
    return np.array(X), np.array(y)

look_back = 12
X_train, y_train = create_dataset(train_data, look_back)
X_test, y_test = create_dataset(test_data, look_back)

# Reshape input to be [samples, time steps, features] for LSTM
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

With the data ready we can start building our model! With two LSTM layers to capture more complex patterns in time series data. The first LSTM layer has 128 units and returns sequences to pass the data to the second LSTM layer, which has 64 units and outputs the final prediction. Dropout layers are added after each LSTM to prevent overfitting by randomly ignoring some neurons during training.

```python
# Model
model = Sequential([
    Input(shape=(X_train.shape[1], X_train.shape[2])),
    LSTM(128, activation='relu', return_sequences=True),
    Dropout(0.2),
    LSTM(64, activation='relu', return_sequences=False),
    Dropout(0.2),
    Dense(1)
])


# Compile the model
model.compile(optimizer='adam', loss='mse')
```

```python
# Define early stopping with validation loss monitoring
early_stopping = EarlyStopping(monitor='val_loss', patience=40, restore_best_weights=True)

# Reduce learning rate when validation loss stops improving
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.8, patience=5, min_lr=1e-4)

history = model.fit(
    X_train, y_train,
    epochs=200,
    batch_size=16,
    validation_data=(X_test, y_test),
    verbose=1,
    callbacks=[early_stopping, lr_scheduler]
)
```

Why Evaluate and Visualize?
Evaluating the model on test data gives us an idea of its performance on unseen data. Visualization of the predicted vs. actual values provides insights into how well the model captures the trends and seasonality.

```python
# Make predictions
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)

# Inverse transform predictions and true values to original scale
train_pred = scaler.inverse_transform(train_pred)
test_pred = scaler.inverse_transform(test_pred)
y_train_orig = scaler.inverse_transform([y_train])
y_test_orig = scaler.inverse_transform([y_test])

# Adjust indices for train and test predictions to account for look_back
train_index = data.index[look_back:train_size]
test_index = data.index[train_size + look_back:]

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(train_index, y_train_orig.flatten(), label='Train Actual')
plt.plot(test_index, y_test_orig.flatten(), label='Test Actual')
plt.plot(train_index, train_pred.flatten(), label='Train Predicted', linestyle='--')
plt.plot(test_index, test_pred.flatten(), label='Test Predicted', linestyle='--')
plt.legend()
plt.title('LSTM Model Predictions')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.grid()
plt.show()

```

Evaluating a time series model involves not just visualizing predictions but also calculating metrics that quantify the accuracy of the model. Common evaluation metrics for regression models, including time series forecasting, are:

Mean Absolute Error (MAE): Measures the average magnitude of the errors in the predictions, without considering their direction. 

Root Mean Squared Error (RMSE): Gives more weight to larger errors, making it useful for detecting significant prediction issues.

R² Score: Indicates how well the predictions explain the variance of the data. A value closer to 1 indicates better performance.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Calculate metrics
train_mae = mean_absolute_error(y_train_orig.flatten(), train_pred.flatten())
test_mae = mean_absolute_error(y_test_orig.flatten(), test_pred.flatten())

train_rmse = np.sqrt(mean_squared_error(y_train_orig.flatten(), train_pred.flatten()))
test_rmse = np.sqrt(mean_squared_error(y_test_orig.flatten(), test_pred.flatten()))

train_r2 = r2_score(y_train_orig.flatten(), train_pred.flatten())
test_r2 = r2_score(y_test_orig.flatten(), test_pred.flatten())

# Display metrics
print(f"Train MAE: {train_mae:.2f}, Test MAE: {test_mae:.2f}")
print(f"Train RMSE: {train_rmse:.2f}, Test RMSE: {test_rmse:.2f}")
print(f"Train R²: {train_r2:.2f}, Test R²: {test_r2:.2f}")
```

The evaluation metrics indicate the model's performance on both the training and test sets. Let’s analyze these results:

1. Mean Absolute Error (MAE):
Train MAE: 12.91: On average, the model’s predictions are off by about 12.91 units on the training set, indicating a relatively good fit to the training data.
Test MAE: 18.85: The slightly higher error on the test set, at 18.85, suggests that the model's predictions are less accurate on unseen data, which is completly normal. While still reasonable, this indicates that the model is not generalizing as well as it does on the training data, this can be improved considering regularization techniques such as increasing early stopping patience or adjusting the model’s complexity.

3. Root Mean Squared Error (RMSE):
Train RMSE: 16.24: The RMSE on the training set is 16.24, which reflects the average magnitude of error, with higher penalties for larger errors. The relatively low value indicates a good fit on the training data, though it’s still higher than the MAE, suggesting that there are some larger errors in the predictions.
Test RMSE: 23.37: The test RMSE is higher, at 23.37, showing that the model struggles more with larger prediction errors on the test set. The increase in RMSE indicates that the model is having more difficulty predicting unseen data accurately, as we have seen also with the MAE values.

5. R² Score:
Train R²: 0.97: The R² score of 0.97 on the training set indicates that the model explains 97% of the variance in the training data, which suggests an excellent fit on the training set.
Test R²: 0.9: The drop in R² to 0.9 on the test set shows that the model is still explaining a large portion of the variance on unseen data. While the R² on the test set is still very good, the slight decrease compared to the training set suggests a small degree of overfitting.

The model demonstrates strong performance on both the training and test sets. With low MAE, RMSE, and high R² on both datasets, it is clear that the model captures the underlying patterns well. However, there is a slight drop in performance on the test set, suggesting that the model is slightly overfitting to the training data.
Nonetheless, the results indicate that the model is performing well and is capable of making accurate predictions on new data.
